```python
x = int(input())
is_prime = True
if x < 2:
    is_prime = False
else:
    for div in range(2, x):
        if x % div == 0:
            is_prime = False
            break

if is_prime:
    print(x, "is prime")
else:
    print(x, "is not prime")
```
