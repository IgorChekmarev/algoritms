Поиск НОД для 2 чисел

Пусть $a >= b$. Рассмотрим тривиальный случай $а = b*c$. Тогда ответом является b. 
Пусть $a >= b$; $a != b*c$. Пусть $a = a`*x$, $b` =  b*x$. Тогда $a - b = x*(a` - b`)$. Можно сказать, что $a = c`*b + a%b$. 


```python
a = int(input())
b = int(input())

while b != 0:
    a, b = b, a % b

print(a)
```

Далее - римская. Здесь проблема в количестве символов: на каждый порядок - 2 новых символа.
Хочется перейти к такой записи числа, где ограниченное числа символов
Переход к стандартной десятичной системе (позиционной). Например, в числе 3728 цифра 7 означает не 7, а 700. То есть 3728 = 3000 + 700 + 20 + 8 = 3*10^3 + 7*10^2 + 2*10^1 + 8*10^0
Но мы не обязаны брать за основание 10 - можно взять любое число, больше 1. Пример - двоичная система счисления. 110101 = 1*2^5 + 1*2^4 + 1*2^2 + 1 = 32 + 16 + 4 + 1 = 53.
Если работаем в системе с основанием больше 10, то добавляем символы из заглавных латинских букв. Пример - 16-ричная система счисления: 1, 2, 3, 4,5 , 6, 7, 8, 9, A, B, C, D, E, F.

Быстрые переводы между системами счислениия

Если у одной системы основание n, а у другой - n^m, то переводы между такими системами облегчены. Пример - двоичная и шестнадцатиричная.

В этом случае m разрядов системы с меньшим основанием конвертируются в один разряд системы с большим основанием.

0010 1110 1111 = 2EF

```python
#для систем счисления до 16 включительно
x = int(input())
base = int(input())

res = ''
alphabet = "0123456789ABCDEF"
while x > 0:
    res = alphabet[x % base] + res
    x //= base

print(res)
```

```python
s = input()
base = int(input())
alphabet = "0123456789ABCDEF"

res = 0
for i in range(len(s)):
    res *= base 
    res += alphabet.index(s[i])

print(res)
```
